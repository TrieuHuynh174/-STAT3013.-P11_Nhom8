```python
import pandas as pd
import scipy.stats as stats

# Đọc dữ liệu từ file CSV
data = pd.read_csv('Get_data/Tesla_Close_2019_2024.csv', parse_dates=['Date'], index_col='Date')

# Tính các giá trị thống kê
mean_close = data['Close'].mean()
median_close = data['Close'].median()
mode_close = data['Close'].mode()[0]
range_close = data['Close'].max() - data['Close'].min()
variance_close = data['Close'].var()
std_dev_close = data['Close'].std()
q1 = data['Close'].quantile(0.25)  
q2 = data['Close'].quantile(0.50) 
q3 = data['Close'].quantile(0.75)
# In kết quả
print(f"Mean (Average) stock price: {mean_close:.2f}")
print(f"Median (Middle) stock price: {median_close:.2f}")
print(f"Mode (Most frequent) stock price: {mode_close:.2f}")
print(f"Range of stock prices: {range_close:.2f}")
print(f"Variance of stock prices: {variance_close:.2f}")
print(f"Standard Deviation of stock prices: {std_dev_close:.2f}")
print(f"25th percentile (Q1): {q1:.2f}")
print(f"50th percentile (Q2 - Median): {q2:.2f}")
print(f"75th percentile (Q3): {q3:.2f}")
```

```python
from matplotlib.dates import DateFormatter

# Vẽ biểu đồ giá đóng cửa
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.title('Tesla Stock Close Price (2019-2024)')
plt.xlabel('Year')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid()

# Định dạng trục x chỉ hiển thị năm
date_format = DateFormatter("%Y")
plt.gca().xaxis.set_major_formatter(date_format)

# Hiển thị biểu đồ
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Close'], bins=50, edgecolor='black', color='skyblue')
plt.title('Histogram of Tesla Stock Prices')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Close'], color='lightgreen')
plt.title('Box Plot of Tesla Stock Prices')
plt.xlabel('Closing Price (USD)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Volume'], data['Close'], alpha=0.5, color='purple')
plt.title('Scatter Plot of Tesla Stock Prices vs Volume')
plt.xlabel('Trading Volume')
plt.ylabel('Closing Price (USD)')
plt.grid()
plt.show()
```

```python
from scipy.stats import kurtosis
import seaborn as sns
import matplotlib.pyplot as plt

# Tính Kurtosis cho Tesla
kurtosis_tesla = kurtosis(data['Close'], fisher=True)

# Tính Skewness cho Tesla
mean_price_tesla = data['Close'].mean()
median_price_tesla = data['Close'].median()
std_tesla = data['Close'].std()
skewness_tesla = (3 * (mean_price_tesla - median_price_tesla)) / std_tesla

# Tạo biểu đồ phân phối với Kurtosis và Skewness
plt.figure(figsize=(10, 6))

sns.histplot(data['Close'], kde=True, color='blue', bins=50, alpha=0.6)
plt.axvline(mean_price_tesla, color='red', linestyle='--', label=f'Mean: {mean_price_tesla:.2f}')
plt.axvline(median_price_tesla, color='green', linestyle='-', label=f'Median: {median_price_tesla:.2f}')

# Lấy giới hạn của trục y để định vị văn bản
y_max = plt.gca().get_ylim()[1]

# Hiển thị Kurtosis và Skewness tại vị trí dễ đọc
plt.text(mean_price_tesla + 0.5 * std_tesla, y_max * 0.85, f'Kurtosis: {kurtosis_tesla:.2f}', color='black', fontsize=10)
plt.text(mean_price_tesla + 0.5 * std_tesla, y_max * 0.75, f'Skewness: {skewness_tesla:.2f}', color='black', fontsize=10)

plt.title('Tesla Close Price Distribution (Skewness & Kurtosis)')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
