```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from the CSV file
data = pd.read_csv('Get_data/Tesla_Close_2019_2024.csv', parse_dates=['Date'], index_col='Date')
```

```python
mean_close = data['Close'].mean()
median_close = data['Close'].median()
modes_close = data['Close'].mode()

print(f"Mean (Average) stock price: {mean_close:.2f}")
print(f"Median (Middle) stock price: {median_close:.2f}")

# Kiểm tra số lượng mode
if len(modes_close) == 1:
    print(f"Mode (Most frequent) stock price: {modes_close[0]:.2f}")
else:
    print(f"Multiple modes found: {', '.join([f'{mode:.2f}' for mode in modes_close])}")
```

```python
range_close = data['Close'].max() - data['Close'].min()
variance_close = data['Close'].var()
std_dev_close = data['Close'].std()

print(f"Range of stock prices: {range_close:.2f}")
print(f"Variance of stock prices: {variance_close:.2f}")
print(f"Standard Deviation of stock prices: {std_dev_close:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
df = pd.read_csv('Get_data/Tesla_Close_2019_2024.csv', parse_dates=['Date'], index_col='Date')


# Tính toán MA(50) và MA(200)
df['MA50'] = df['Close'].rolling(window=50).mean()
df['MA200'] = df['Close'].rolling(window=200).mean()

# Vẽ biểu đồ giá cổ phiếu với trục y bên trái
fig, ax1 = plt.subplots(figsize=(14, 7))

# Vẽ đường giá đóng cửa
ax1.plot(df.index, df['Close'], color='blue', label='Close Price', linewidth=2)

# Vẽ đường MA(50)
ax1.plot(df.index, df['MA50'], color='green', label='MA (50)', linewidth=1.5)

# Vẽ đường MA(200)
ax1.plot(df.index, df['MA200'], color='red', label='MA (200)', linewidth=1.5)

ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price (USD)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Tesla Stock Price, MA(50), MA(200), and Trading Volume (2019-2024)')

# Tạo trục y thứ hai cho Volume
ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color='orange', alpha=0.6, label='Volume')
ax2.set_ylabel('Volume', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Hiển thị chú thích
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Đọc dữ liệu từ file CSV
data = pd.read_csv('Get_data/Tesla_Close_2019_2024.csv', parse_dates=['Date'], index_col='Date')

# Tính các giá trị thống kê
mean_close = data['Close'].mean()
median_close = data['Close'].median()
mode_close = data['Close'].mode()[0]
range_close = data['Close'].max() - data['Close'].min()
variance_close = data['Close'].var()
std_dev_close = data['Close'].std()
kurtosis_close = stats.kurtosis(data['Close'])
skewness_close = stats.skew(data['Close'])
q1 = data['Close'].quantile(0.25)  
q2 = data['Close'].quantile(0.50) 
q3 = data['Close'].quantile(0.75)
# In kết quả
print(f"Mean (Average) stock price: {mean_close:.2f}")
print(f"Median (Middle) stock price: {median_close:.2f}")
print(f"Mode (Most frequent) stock price: {mode_close:.2f}")
print(f"Range of stock prices: {range_close:.2f}")
print(f"Variance of stock prices: {variance_close:.2f}")
print(f"Standard Deviation of stock prices: {std_dev_close:.2f}")
print(f"Kurtosis of stock prices: {kurtosis_close:.2f}")
print(f"Skewness of stock prices: {skewness_close:.2f}")
print(f"25th percentile (Q1): {q1:.2f}")
print(f"50th percentile (Q2 - Median): {q2:.2f}")
print(f"75th percentile (Q3): {q3:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('Get_data/Tesla_Close_2019_2024.csv', parse_dates=['Date'], index_col='Date')

# Vẽ biểu đồ giá đóng cửa
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.title('Tesla Stock Close Price (2019-2024)')
plt.xlabel('Year')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid()

# Định dạng trục x chỉ hiển thị năm
date_format = DateFormatter("%Y")
plt.gca().xaxis.set_major_formatter(date_format)

# Hiển thị biểu đồ
plt.show()
```
