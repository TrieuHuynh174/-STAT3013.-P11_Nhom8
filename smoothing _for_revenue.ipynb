```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'Daily_Revenue_2019_2024.csv'
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'])

data['Smoothed Revenue'] = data['Daily Revenue'].rolling(window=30).mean()

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Daily Revenue'], label='Original Daily Revenue', alpha=0.4, color='blue')
plt.plot(data['Date'], data['Smoothed Revenue'], label='30-Day Moving Average', color='red')
plt.title("Daily Revenue (2019-2024)")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.show()

```
