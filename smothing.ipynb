```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Loading the dataset
file_path = 'C:\\Users\\phamp\\Downloads\\Daily_Sales_2019_2024.csv'
data = pd.read_csv(file_path)

data.index += 1
```

```python
# Convert 'Date' column to datetime format for better handling and visualization
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Re-plotting with updated 'Date' format
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Daily Tesla Sales'], label='Daily Tesla Sales', alpha=0.3)
plt.plot(data['Date'], data['Exponential Smoothing'], label='Exponential Smoothing', color='red')
plt.xlabel('Date')
plt.ylabel('Tesla Sales')
plt.title('Daily Tesla Sales with Exponential Smoothing')
plt.legend()
plt.xticks(rotation=45)  # Rotate dates for better readability
plt.tight_layout()
plt.show()
```
