```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python

tesla_daily_data = pd.read_csv('Daily_Sales_2019_2024.csv')

tesla_daily_data.index += 1
```

```python
tesla_daily_data['Date'] = pd.to_datetime(tesla_daily_data['Date'], errors='coerce')

tesla_daily_data['Year'] = tesla_daily_data['Date'].dt.year

tesla_daily_data['Exponential Smoothing'] = tesla_daily_data['Daily Tesla Sales'].ewm(span=30, adjust=False).mean()

plt.figure(figsize=(12, 6))
plt.plot(tesla_daily_data['Date'], tesla_daily_data['Daily Tesla Sales'], label='Daily Tesla Sales', alpha=0.3)
plt.plot(tesla_daily_data['Date'], tesla_daily_data['Exponential Smoothing'], label='Exponential Smoothing', color='red')
plt.xlabel('Date')
plt.ylabel('Tesla Sales')
plt.title('Daily Tesla Sales with Exponential Smoothing')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
