```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.ndimage import gaussian_filter1d

file_path = 'S&P500_Close_2019_2024.csv'
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

optimal_sigma = 10

data['Smoothed_Close'] = gaussian_filter1d(data['Close'], sigma=optimal_sigma)

plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x=data.index, y='Close', label='Original Closing Price', color='blue')
sns.lineplot(data=data, x=data.index, y='Smoothed_Close', label='Smoothed Closing Price', color='red')
plt.title('S&P 500 Closing Price Smoothing using Gaussian Filter')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid()
plt.show()

```
